## PageRank MRJob:

```python
%%writefile PageRank.py

from mrjob.job import MRJob
from mrjob.step import MRStep
import ast
import sys

class PageRank(MRJob):

    def steps(self):
        JOBCONF_STEP1 = {
            'mapreduce.job.output.key.comparator.class': 'org.apache.hadoop.mapred.lib.KeyFieldBasedComparator',
            'mapreduce.job.reduces': 16
        }

        return [MRStep(jobconf=JOBCONF_STEP1,
                       mapper=self.mapper,
                       reducer=self.reducer),
                MRStep(reducer=self.reducer_pagerank)
                ]



    def configure_options(self):
        super(PageRank, self).configure_options()
        self.add_passthrough_option('--dampingfactor', default='0.85', type=str, help='damping factor')
        self.add_passthrough_option('--numtotalnodes', default=None, type=str, help='total number of nodes in the graph')


    def mapper(self, _, line):
        node = None
        adjacency_list = None
        initial_page_rank = 0.0
        neighbors = 0
        if line!= "":
            node, graph_pr_out_links = line.strip().split('\t')
            node =  node.replace('"', '')

            adjacency_list, initial_page_rank = map(ast.literal_eval, graph_pr_out_links.strip("\"").strip('"').strip("'").split('|'))
            neighbors = len(adjacency_list)

            if neighbors> 0:
                for key in adjacency_list.keys():
                    yield (key),  (1.0* initial_page_rank /neighbors)

            if neighbors== 0:
                yield '**dangling', (1.0* initial_page_rank)

            yield (node), (adjacency_list)


    def reducer(self, key, values):
        node = None
        adjacency_list = {}
        total = 0.0
        node =  key.strip("'")
        #values = [v  for v in values]
        redistribute_mass = 0.0


        if node == '**dangling':
            redistribute_mass= sum([float(v) for v in values])
            for i in range(1, int(self.options.numtotalnodes) + 1):
                yield str(i), redistribute_mass

        else:
            for pr_out_link in values:
                if type(pr_out_link) == float:
                    total = total + float(pr_out_link)

                elif type(pr_out_link) == dict:
                    adjacency_list = pr_out_link

            yield node , (str(adjacency_list) + '|' + str(total) )



    def reducer_pagerank(self, node, page_rank_adj_lists):
        lost_mass = 0.0
        page_rank = 0.0
        adj_lists = {}

        alpha = 1 - float(self.options.dampingfactor)
        total_nodes = float(self.options.numtotalnodes)

        for page_rank_adj_list in page_rank_adj_lists:
            if type(page_rank_adj_list) == float:
                lost_mass = page_rank_adj_list
            else:
                adj_lists, page_rank = map(ast.literal_eval, page_rank_adj_list.strip("\"").split('|'))

        page_rank_updated = alpha * (1/total_nodes) + (1-alpha) * (float(lost_mass)/total_nodes + float(page_rank))

        yield node, str(adj_lists) + "|" + str(round(page_rank_updated, 6))


if __name__ == '__main__':
    PageRank.run()
```

```python
%%writefile PageRank_driver.py
from PageRank import PageRank
from mrjob.job import MRJob
from mrjob.step import MRStep
import sys
import ast
import subprocess

INPUT_GRAPH = sys.argv[1]
DAMPING_FACTOR = sys.argv[2]
TOTAL_NODES = sys.argv[3]
MAX_ITERATION = int(sys.argv[4])
MODE = sys.argv[5]

num_interation = 0


mr_job = PageRank(args=[INPUT_GRAPH, '--dampingfactor', DAMPING_FACTOR, '--numtotalnodes', TOTAL_NODES,  '-r', 'local' ])
while (num_interation<MAX_ITERATION):
    with mr_job.make_runner() as runner:
        num_interation += 1
        print("Iteration: ", num_interation)
        runner.run()
        f = open(INPUT_GRAPH, 'w+')
        for line in runner.stream_output():
            # if num_interation%5 == 0:
              # print(line.decode("utf-8"))
              # print()
            f.writelines(line.decode("utf-8"))

    f.close()
```

#**Varying Damping Factor**

```python
%%writefile explore_pagerank_teleportation_driver.py

from PageRank import PageRank
from mrjob.job import MRJob
from mrjob.step import MRStep
import sys
import ast
import subprocess
from shutil import copyfile

SOURCE_GRAPH = './data_updated.txt'
DAMPING_FACTOR = [0, 0.25, 0.5, 0.75, 0.85, 1]
TOTAL_NODES = 11
MAX_ITERATION = 30
MODE = 'local'

for d in DAMPING_FACTOR:
    print('Started job with damping factor:', d)
    INPUT_GRAPH = './data_input_with_'+str(d)+'.txt'
    copyfile(SOURCE_GRAPH, INPUT_GRAPH)

    mr_job = PageRank(args=[INPUT_GRAPH, '--dampingfactor', str(d), '--numtotalnodes', str(TOTAL_NODES),  '-r', 'local' ])
    num_interation = 0
    while(num_interation<MAX_ITERATION):

        with mr_job.make_runner() as runner:

            num_interation += 1
            runner.run()

            f = open(INPUT_GRAPH, 'w+')
            for line in runner.stream_output():
                f.writelines(line.decode("utf-8"))

        f.close()
```

```python
%matplotlib inline
import networkx as nx
import ast
from matplotlib import pyplot as plt

def plot_graph(graph, pagerank, damping, labels=None):
    plt.figure(figsize=(5, 5))  # Reduced image size to 1/3

    Directed_Graph = nx.DiGraph()

    for edge in graph:
        Directed_Graph.add_edge(edge[0], edge[1])

    # Normalize PageRank values for scaling
    max_pr = max(pagerank.values())
    min_pr = min(pagerank.values())
    norm_pagerank = {}
    if (max_pr - min_pr) != 0:
        norm_pagerank = {node: (value - min_pr) / (max_pr - min_pr) for node, value in pagerank.items()}

    # Node size based on normalized PageRank
    node_sizes = [2000 + 5000 * norm_pagerank.get(node, 0) for node in Directed_Graph.nodes()]
    node_colors = [plt.cm.viridis(norm_pagerank.get(node, 0)) for node in Directed_Graph.nodes()]

    # Use a more sophisticated layout with increased spacing (k=0.8 for more space)
    graph_pos = nx.spring_layout(Directed_Graph, seed=42, k=0.8, scale=2)  # Increase `k` for more space between nodes

    # Create labels for nodes with their PageRank percentage
    labels = {node: f"{node}\n{round(pagerank.get(node, 0) * 100, 1)}%" for node in Directed_Graph.nodes()}

    print(f"Network graph with damping factor = {damping}")

    # Draw the nodes with dynamic sizing and coloring
    nx.draw_networkx_nodes(
        Directed_Graph,
        graph_pos,
        node_size=node_sizes,
        node_color=node_colors,
        alpha=1,  # Set opacity to 1 for better visibility
        cmap=plt.cm.viridis
    )

    # Draw edges with slightly transparent lines to avoid clutter
    nx.draw_networkx_edges(
        Directed_Graph,
        graph_pos,
        width=2,
        alpha=1,  # Set opacity to 1 for arrows and edges
        edge_color="gray",
        arrows=True,
        arrowsize=20
    )

    # Draw labels for the nodes
    nx.draw_networkx_labels(
        Directed_Graph,
        graph_pos,
        labels=labels,
        font_size=10,
        font_color="black"
    )

    # Add colorbar for PageRank
    sm = plt.cm.ScalarMappable(cmap=plt.cm.viridis, norm=plt.Normalize(vmin=min_pr, vmax=max_pr))
    sm.set_array([])
    cbar = plt.colorbar(sm, shrink=0.8)
    cbar.set_label("PageRank Value", rotation=270, labelpad=15)

    # Set the title and remove axis
    plt.title(f"Graph Visualization with Damping Factor = {damping}", fontsize=16, fontweight="bold")
    plt.axis("off")

    # Adjust padding to avoid cutting off nodes
    plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1)

    plt.tight_layout()
    plt.show()

# Example Usage with Inputs
DAMPING_FACTOR = [0, 0.25, 0.5, 0.75, 0.85, 1]
graph = []
pagerank = {}
node_dict = {1: 'A', 2: 'B', 3: 'C', 4: 'D', 5: 'E', 6: 'F'}

for damping in DAMPING_FACTOR:
    input_file = "wikipedia_example_input_with_" + str(damping) + ".txt"
    with open(input_file, "r") as f:
        for line in f.readlines():
            line = line.strip()
            node, pagerank_out_links = line.split('\t')
            node = int(node.strip('"'))
            if node not in node_dict:
                continue  # Skip nodes outside the desired range (1 to 6)
            out_links, pr = map(ast.literal_eval, pagerank_out_links.strip('"').split('|'))
            graph.extend([(node_dict.get(int(node), str(node)), node_dict.get(int(n), str(n))) for n in out_links.keys() if int(n) in node_dict])
            pagerank[node_dict.get(int(node), str(node))] = pr
        plot_graph(graph, pagerank, damping)
        graph.clear()  # Clear the graph for the next iteration
```
